```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
spotify_df = pd.read_csv('song_data(pop-artist).csv')
```

```python
spotify_df.head(30)
```

```python
reformed_df = spotify_df.drop(columns = ['id', 'title', 'all_artists'])
reformed_df = reformed_df.iloc[: , 1:]
reformed_df = reformed_df.drop(columns = ['release_date', 'key', 'mode', 'duration_ms', 'time_signature'])
# remove rows by filtering
reformed_df = reformed_df[reformed_df['popularity_song'] != 0]
# display the dataframe
reformed_df
```

```python
reformed_df.describe()
```

```python
popularity_target = []
for i in reformed_df.popularity_song:
    if i > 64:
        popularity_target.append(1)
    else:
        popularity_target.append(0)

reformed_df['popularity_target'] = popularity_target
reformed_df.head()
```

```python
X = reformed_df.drop(columns = ['popularity_song','popularity_target'])
y = reformed_df.popularity_target
X.head()
        
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 100)
scaler = StandardScaler()
```

```python
# from imblearn.under_sampling import RandomUnderSampler
# from collections import Counter
# ros = RandomUnderSampler(random_state=1)
# X_train, y_train = ros.fit_resample(X_train, y_train)
# Counter(y_train)
```

```python
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train_scaled, y_train)
```

```python
y_pred = rf.predict(X_test_scaled)
```

```python
y_pred
```

```python
# Calculating the confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, y_pred)
```

```python
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, y_pred))
```
